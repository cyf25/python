```python
cust=pd.read_sql_query("select CustomerId,FirstName,LastName from customers",con)
invoice=pd.read_sql_query("select InvoiceId,CustomerId from invoices",con)
ii=pd.read_sql_query("select InvoiceLineId,InvoiceId,UnitPrice,Quantity from invoice_items",con)
```

```python
cust
```

```python
invoice
```

```python
ii
```

```python
cust_inv=cust.merge(invoice,on='CustomerId').merge(ii,on='InvoiceId')
cust_inv
```

# 计算用户每笔消费的总金额
# DataFrame的assign方法 创建新列


```python
total=cust_inv['Quantity']*cust_inv['UnitPrice']
cust_inv=cust_inv.assign(Total=total)#给cust_inv新加一列Total
cust_inv.head()
```

```python
cust_inv.groupby(['CustomerId','FirstName','LastName'])['Total'].sum().sort_values(ascending=False).head()
```

```python
stocks_2016=pd.read_csv(r'D:\BaiduNetdiskDownload\day4\stocks_2016.csv')
stocks_2017=pd.read_csv(r'D:\BaiduNetdiskDownload\day4\stocks_2017.csv')
stocks_2018=pd.read_csv(r'D:\BaiduNetdiskDownload\day4\stocks_2018.csv')
```

# 如果合并的两个数据有相同的列名，需要通过lsuffix，和rsuffix，指定合并后的列名的后缀

```python
stocks_2016.join(stocks_2017,lsuffix='_2016',rsuffix='_2017',how='outer')
```

# 将两个DataFrame的Symbol设置为行索引，再次join数据


```python
stocks_2016.set_index('Symbol').join(stocks_2017.set_index('Symbol'),lsuffix='_2016',rsuffix='_2017')
```

# 将一个DataFrame的Symbol列设置为行索引，与另一个DataFrame的Symbol列进行join


```python
stocks_2016.join(stocks_2017.set_index('Symbol'),lsuffix='_2016',rsuffix='_2017',on='Symbol') 
```
