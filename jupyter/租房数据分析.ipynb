# 最受欢迎的户型

# 房子的平均房租价格

```python
house_data.loc[:,'平均房租']=house_data['价格']/house_data['面积']
house_data.loc[:,'平均房租'].mean()
# house_data.loc[:,'平均房租'].median()
```

# 最热门小区

```python
house_data[['地址','看房人数']].groupby('地址').sum().sort_values(by='看房人数',ascending=False).head(10)
```

```python
house_data[['地址','看房人数']].groupby('地址').count().sort_values(by='看房人数',ascending=False)
```

# agg 聚合 {'列名':'聚合方式'},实现不同字段，用不同的聚合方法

```python
house_data.groupby('户型').agg({'看房人数':'sum','价格':'mean'}).sort_values(by='看房人数',ascending=False).head(10)
```
