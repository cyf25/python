##查看每年评分最高的电影
1/先对数据按照年份进行降序
2/相同年份的话再按照评分进行降序
3/去重，获取第一个数据，得到每年评分最高的电影

```python
movie2=movie[['movie_title','imdb_score','title_year']]
movie2.sort_values('title_year',ascending=False)
```

```python
movie3=movie2.sort_values(['title_year','imdb_score'],ascending=False)
movie3.head()
```

```python
movie_top_year=movie3.drop_duplicates('title_year')
movie_top_year.head()
```
